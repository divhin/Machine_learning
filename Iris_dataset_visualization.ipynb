## Dataset Description: 
Iris dataset from Sk_learn <br>
The dataset contains <br>
target : Iris Flower Species <br>
features: 'sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'

## Visualizing target vs features through scatterplots

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
from sklearn.datasets import load_iris
```

```python
iris_dataset = load_iris()
```

```python
iris_dataset.keys()
```

```python
iris_dataset['feature_names']
```

```python
data =np.array(iris_dataset['data'])
```

```python
feature_data=data.T
target_data=np.array(iris_dataset['target'])
target_data
feature_data.shape
```

```python
no_of_features=feature_data.shape[0]
fig, ax = plt.subplots(no_of_features,no_of_features,figsize=(16,16))

for i in range(no_of_features):
    
    for j in range(no_of_features):
        feature_with_types=np.array((feature_data[i],feature_data[j],target_data)).T
        type_0=feature_with_types[feature_with_types[:,2] == 0]
        type_1=feature_with_types[feature_with_types[:,2] == 1]
        type_2=feature_with_types[feature_with_types[:,2] == 2]
        
        if i == j:
            ax[i,j].hist(feature_data[j],facecolor = 'b',edgecolor='k')
            #ax[i,j].yaxis.set_ticklables([])
            
        else:
            ax[i,j].scatter(type_0.T[1],type_0.T[0],c = 'b',edgecolors='k',label = 'setosa')
            ax[i,j].scatter(type_1.T[1],type_1.T[0],c = 'r',edgecolors = 'k',label = 'versicolor')
            ax[i,j].scatter(type_2.T[1],type_2.T[0],c = '#90EE90',edgecolors='k', label = 'virginica')
            #ax[i,j].legend(ncol = 3,loc = 'upper center')
            
#plt.subplots_adjust(wspace=0,hspace=0)
labels = iris_dataset['feature_names']

for i in range(ax.shape[0]):
    for j in range(ax.shape[1]):
        ax[i,j].set_xlabel(labels[j])
        ax[i,j].set_ylabel(labels[i])

red_dot = mpatches.CirclePolygon(color='red', label='setosa',xy=(1,1),radius=1)
blue_dot = mpatches.CirclePolygon(color='blue', label='versicolor',xy=(1,2),radius=1)
lightgreen_dot = mpatches.CirclePolygon(color='#90EE90', label='virginica',xy=(1,3),radius=1)

plt.legend(handles=[red_dot,blue_dot,lightgreen_dot],bbox_to_anchor = (-0.75,-0.18), ncol=3, borderaxespad=0.)
        
fig.text(0.35,0.9,'Iris Dataset Feature Plot',fontweight = 'bold', fontsize=30)

#ax.legend(bbox_to_anchor=(1.05, 0), loc='lower left', borderaxespad=0.)
plt.savefig('task32.png')
plt.show()
            
```

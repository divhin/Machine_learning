## Visualization of Spurious Correlations
Total Revenue Generated by Arcades <br> **Correlates with** <br> Number of Computer Science Doctorates Awarded in the US

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
revenue_data=pd.read_csv('revenue.txt',sep=',')
degrees_data=pd.read_csv('degrees.txt',sep=',')
```

```python
revenue=np.array(revenue_data['revenue'])
degrees=np.array(degrees_data['degrees'])
year=np.array(revenue_data['Year'])
```

```python
fig = plt.figure(figsize=(12,6))
ax1=plt.gca()
line1,=ax1.plot(year,degrees,'-o',lw=2,c='black',drawstyle='default')
ax1.set_xlabel('Year')
ax1.set_xticklabels(list(year))
ax2=ax1.twinx()
line2,=ax2.plot(year,revenue,'-s',lw=2,c='#800000',drawstyle='default')
ax1.set_ylabel('Computer Science Doctorates (degrees)')
ax2.set_ylabel('Arcade Revenue ($ billion)')
plt.title('Total Revenue Generated by Arcades \n Correlates with \n Number of Computer Science Doctorates Awarded in the US',fontsize=14, fontweight='bold')

plt.legend((line1,line2),("Computer Science Doctorates","Arcades Revenue"))
plt.grid(which='major', axis='y')
plt.savefig('task31.png')
plt.show()
```
